# Librerías

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Datos simulados

```python
n=100 # numero de observaciones

x=np.random.randint(10,high=100,size=n) # Variable independiente

beta=(100,0.5) # Coeficientes b0 y b1
sigma=10 # Desviación típica del error

err=np.random.randn(n)*sigma # Vector de errores

y=beta[0]+beta[1]*x+err # Vector de la variable dependiente

# Gráfico
fig, ax = plt.subplots()
ax.plot(x,y,'o')
ax.set_title('Grafico xy de los puntos del dataset')
fig.show()
```

# Ajuste del modelo

```python
from sklearn.linear_model import LinearRegression

reg =LinearRegression()
reg.fit(x.reshape(-1,1),y) # juste del modelo

# Calculo de las betas y de r2
t1=('Beta0= ',round(reg.intercept_,2))
t2=('Beta1= ',round(reg.coef_[0],2))
t3=('r2= ',round(reg.score(x.reshape(-1,1), y),2))

# Gráfico
fig, ax = plt.subplots()
ax.plot(x, y,'o',label='Puntos')
ax.plot(x, reg.predict(x.reshape(-1,1)), label='Recta de regresion')
ax.legend()
ax.text(110,160,t1)
ax.text(110,155,t2)
ax.text(110,150,t3)
ax.set_title('Regresión lineal')
fig.show()
```
